```python
#importing all the needed libraries
import pandas as pd
import matplotlib.pyplot as plt

#read the csv file
df=pd.read_csv("/content/Banking/banking_data.csv")
df
```

**Question 1 What is the distribution of age among the clients?**

```python
#calculating the range first
maxAge=df['age'].max()
minAge=df['age'].min()
print("Max Age:",maxAge)
print("Min Age:",minAge)
print(f'the range of age among the clients is {maxAge-minAge} \n')

#making a histogram for better visualtion of the distribution of age
plt.hist(df['age'], bins=10, edgecolor='black')
plt.xlabel('age')
plt.ylabel('number of clients')
plt.title("distribution of age among the clients")
plt.show()
```

**Question 2 How does the job type vary among the clients?**

```python
#different types of jobs
job_type=df['job'].unique()
print("all job types are shown below:")
for job in job_type:
  print(job)

print('\n')
#number of clients in each type of job
job_count=df['job'].value_counts()
print(job_count)

print('\n')
#ploting a bar plot for visualization
plt.figure(figsize=(12, 7))
plt.bar(job_count.index, job_count)
plt.xlabel('job type')
plt.ylabel('number of clients')
plt.title('Job type distribution of clients')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**Question 3 What is the marital status distribution of the clients?**

**Question 4 What is the level of education among the clients?**

**Question 5 What proportion of clients have credit in default?**

```python
default_card=df['default'].value_counts()
print(default_card)

# Calculate the proportions
total_clients = len(df)
print('total number of clients:',total_clients)
default_proportion = default_card / total_clients

# Create a pie chart
plt.pie(default_proportion, labels=default_proportion.index, autopct='%1.1f%%')
plt.title('Proportion of Clients with Credit in Default')
plt.axis('equal')
plt.show()

# Print the proportions
print("Proportion of Clients with Credit in Default:")
print(default_proportion)
```

**Question 14 What is the distribution of the number of days passed since the client was last contacted from a previous campaign?**


```python
#get the count of clients who were previously not connected
pdays_not_contacted_count = (df['pdays'] == -1).sum()
print(f"Number of clients not previously contacted: {pdays_not_contacted_count} \n")

# Plot a histogram of the 'pdays' column, including the '-1' value.
plt.figure(figsize=(10, 6))
df['pdays'].plot(kind='hist', bins=30, edgecolor='black')

# Set the x-axis limits to focus on the positive pdays values,
# effectively hiding the large bar at -1.
plt.xlim(0, df['pdays'].max())

plt.title('Distribution of Days Passed Since Last Contact (excluding -1)')
plt.xlabel('Days Passed')
plt.ylabel('Number of Clients')
plt.show()

# Print some basic statistics for all pdays values
print("\nDays Since Previous Contact Statistics (all values):")
print(df['pdays'].describe())
```

**Question 18 Are there any correlations between different attributes and the likelihood of subscribing to a term deposit?**

```python
import seaborn as sns

#we are converting the target(y) into a numerical value from catagorial value 'yes' and 'no'
df['y_encoded']=df['y'].map({'yes':1, 'no':0})

#we are selecting the columns which has numerical values only
numerical_df = df[['age', 'balance', 'duration', 'campaign', 'pdays', 'previous', 'y_encoded']]

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Create a heatmap to visualize the correlations
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Attributes and Term Deposit Subscription')
plt.show()

# Print the correlations with the target variable 'y_encoded'
print("Correlation of numerical attributes with term deposit subscription:")
print(correlation_matrix['y_encoded'].sort_values(ascending=False))
```
